```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2 . imread ('my.png')
img_rgb = cv2 . cvtColor ( img , cv2.COLOR_BGR2RGB )
```

```python
plt.figure(figsize=(4,4))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("rgb image")
#conert in gray scale
GRAY_img = cv2.cvtColor(img_rgb , cv2.COLOR_RGB2GRAY)
plt.subplot(1,2,2)
plt.imshow(GRAY_img)
plt.title("grayscale img")
```

***channele sepration***

```python
# rgb
red = img_rgb[:,:,0]
green = img_rgb[:,:,1]
blue = img_rgb[:,:,2]
```

```python
plt.figure(figsize=(4,5))
plt.subplot(1,3,1)
plt.title("red")
plt.imshow(red)

plt.subplot(1,3,2)
plt.title("green")
plt.imshow(green)

plt.subplot(1,3,3)
plt.title("blue")
plt.imshow(blue)
```

```python
#rotate iamge to 45 degree
import imutils
rotated_img = imutils.rotate_bound(img_rgb, 45)
plt.figure(figsize=(4,4))
plt.title('rgb images')
plt.imshow(rotated_img)
```

***image resize***

```python
resized_img = cv2.resize(img_rgb,(0,0), fx=0.5, fy=0.5, interpolation= cv2.INTER_LINEAR)
plt.imshow(resized_img)
```

```python
#plt historical diagram
plt.hist(GRAY_img, bins=5)  # bins=50 makes it faster
plt.title('Simple Grayscale Histogram')
plt.show
```

***horizental and verticle flip***

```python
H_flip = cv2.flip(img_rgb, 1)
y_flip = cv2.flip(img_rgb, 0)

plt.figure(figsize=(4,5))
plt.subplot(1,3,1)
plt.title("origanl")
plt.imshow(img_rgb)

plt.subplot(1,3,2)
plt.title("horizental")
plt.imshow(H_flip)

plt.subplot(1,3,3)
plt.title("vertical")
plt.imshow(y_flip)
```

***brightness and contrast***

```python
contrast = 3
brightness = 70
bright_img = cv2.convertScaleAbs(img_rgb, alpha= contrast, beta=brightness)
plt.figure(figsize=(4,5))
plt.title("horizental")
plt.imshow(bright_img)
```

*** gaussion blurr ***

```python
blur_img = cv2.GaussianBlur(img_rgb, (15,15),0)
plt.figure(figsize=(4,5))
plt.title("blur")
plt.imshow(blur_img)
```
